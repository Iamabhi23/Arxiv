```python
from umap import UMAP
import hdbscan
```

```python
AS2 = ArchiveSearch(path = '../models')
```

```python
AS2.load_model()
```

```python
umap_args = {'n_neighbors': 15,
            'n_components': 2,
            'metric': 'euclidean'}
umap_model = UMAP(**umap_args).fit(AS2.model.dv.vectors)
```

```python
hdbscan_args = {'min_cluster_size': 15,
                'metric': 'euclidean',
                'cluster_selection_method': 'eom'}
cluster = hdbscan.HDBSCAN(**hdbscan_args).fit(umap_model.embedding_)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.scatter(umap_model.embedding_[:, 0], umap_model.embedding_[:, 1], c=cluster.labels_)
plt.show()
```

```python
import pickle as pkl
```

```python
with open('../models/UMAP', 'wb') as f:
    pkl.dump(umap_model, f)
with open('../models/clusters', 'wb') as f:
    pkl.dump(cluster, f)
```
